```python
import seaborn as sns
import pandas as pd
import warnings
import matplotlib
import numpy
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/house_pricing.csv')
data.describe()
data.info
```

```python
data.head()
```

```python
sns.relplot(x='LotArea', y='SalePrice', data=data)
```

```python
filter_lot_area_and_price = data[data.LotArea<20000][data.SalePrice<500000]
sns.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
```

```python
sns.relplot(x='LotFrontage', y='LotArea', data=data)
```

```python
filter_lot_frontage_and_area = data[data.LotFrontage<200][data.LotArea<100000]
sns.relplot(x='LotFrontage', y='LotArea', data=filter_lot_frontage_and_area)
```
